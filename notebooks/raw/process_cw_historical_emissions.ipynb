```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Source du fichier : https://www.climatewatchdata.org/data-explorer/historical-emissions?historical-emissions-data-sources=pik&historical-emissions-gases=&historical-emissions-regions=&historical-emissions-sectors=&page=1


df_cw_historical_emissions_pik = pd.read_csv('cw/cw_historical_emissions_pik.csv')
display(df_cw_historical_emissions_pik)
```

```python
annees = df_cw_historical_emissions_pik.select_dtypes(include='number').columns

# liste des types de gaz
gases = df_cw_historical_emissions_pik['gas'].unique()

# Créer un dictionnaire pour stocker les sommes des quantités émises par gaz et par année
sommes_par_gas = {}

# Calculer les sommes par gaz et par année
for gas in gases:
    sommes_par_gas[gas] = df_cw_historical_emissions_pik.loc[df_cw_historical_emissions_pik['gas'] == gas, annees].sum()

# Tracer la somme des quantités émises par gaz par année
for gas in gases:
    plt.plot(annees, sommes_par_gas[gas], label=gas)

plt.xlabel('Année')
plt.ylabel('Quantité émise')
plt.title('Évolution des quantités émises par substance')
plt.legend()


plt.show()
```
